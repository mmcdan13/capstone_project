```python
# Dependancies
import pandas as pd
import os
import numpy as np

```

```python
# Loading Files
crime_data_to_load = os.path.join("Resources","2020_crime.csv")
chicago_zip_data_to_load = os.path.join("Resources", "chicago_zip_ward.csv")
chicago_pop_zip_data_to_load = os.path.join("Resources","Chicago_Population_by Zipcode_2020.csv" )
                                     
```

```python
ward_zip_df = pd.read_csv(chicago_zip_data_to_load,header=0, delim_whitespace=False)
ward_zip_df
```

```python
crime_df = pd.read_csv(crime_data_to_load, header=0, delim_whitespace=False)
crime_df
```

```python
crime_df.count()
```

```python
crime_df.dropna()
```

```python
clean_crime_df = crime_df[["ID", "Date", "Block", "Primary Type", "Ward","rankings", "Latitude", "Longitude", ]]
clean_crime_df
```

```python
clean_crime_df.set_index('Ward', inplace = True)
ward_zip_df.set_index('Ward', inplace=True)

crime_zip_df = pd.merge(clean_crime_df, ward_zip_df, left_index = True, right_index = True, how= 'inner')
crime_zip_df.reset_index('Ward', drop=False, inplace=True)

crime_zip_df

```

```python
crime_zip_df.dropna()
```

```python
 
crime_zip_df.to_csv('Resources/Crime_zip.csv')
```

```python
pop_zip_df = pd.read_csv(chicago_pop_zip_data_to_load, header=0, delim_whitespace=False)
pop_zip_df
```

```python


clean_pop_df = pop_zip_df[["ZIP", "Population", "People/Sq.Mile"]]
clean_pop_df
```

```python
pop_zip_df.to_csv('Resources/clean_pop_zip.csv') 
```

```python
crime_zip_df.set_index('ZIP', inplace = True)
clean_pop_df.set_index('ZIP', inplace=True)

crime_zip_pop_df = pd.merge(crime_zip_df, clean_pop_df, left_index = True, right_index = True, how= 'inner')
crime_zip_pop_df.reset_index('ZIP', drop=False, inplace=True)

crime_zip_pop_df
```

```python
crime_zip_pop_df.dropna()
```

```python
crime_zip_pop_df.to_csv('Resources/clean_pop_zip.csv')
```
