```python
# Dependancies
import pandas as pd
import os
import numpy as np

```

```python
# Loading Files -
crime_data_to_load = os.path.join("Resources","2020_crime.csv")
chicago_zip_data_to_load = os.path.join("Resources", "chicago_zip_ward.csv")
chicago_pop_zip_data_to_load = os.path.join("Resources","Chicago_Population_by_Zipcode_2020.csv" )
                                     
```

```python
# reading in ward/zip data into dataframe for merging
ward_zip_df = pd.read_csv(chicago_zip_data_to_load,header=0, delim_whitespace=False)
ward_zip_df
```

```python
#Reading in crime data into dataframe
crime_df = pd.read_csv(crime_data_to_load, header=0, delim_whitespace=False)
crime_df
```

```python
#exploring data
crime_df.count()
```

```python
#cleaning data down to columns needed for analysis
clean_crime_df = crime_df[["Primary Type", "Ward","rankings", "Latitude", "Longitude", "Arrest", "Domestic" ]]
clean_crime_df
```

```python
# merging zip data from ward/zip data into the crime data on index Ward
clean_crime_df.set_index('Ward', inplace = True)
ward_zip_df.set_index('Ward', inplace=True)

crime_zip_df = pd.merge(clean_crime_df, ward_zip_df, left_index = True, right_index = True, how= 'inner')
crime_zip_df.reset_index('Ward', drop=False, inplace=True)

crime_zip_df

```

```python
crime_zip_df.to_csv('Resources/Crime_zip_for_model.csv')
```

```python
# read in population/zip data into a df for merging
pop_zip_df = pd.read_csv(chicago_pop_zip_data_to_load, header=0, delim_whitespace=False)
pop_zip_df
```

```python
# clean dataset to columns needed only

clean_pop_rank_df = pop_zip_df[["ZIP", "Population", "People/Sq.Mile", "National Rank"]]
clean_pop_rank_df
```

```python
#write cleaned data to a .csv
clean_pop_rank_df.to_csv('Resources/clean_pop_zip_rank.csv') 
```

```python
# merge population data to crime dataset based on index ZIP
crime_zip_df.set_index('ZIP', inplace = True)
clean_pop_rank_df.set_index('ZIP', inplace=True)

crime_zip_pop_df = pd.merge(crime_zip_df, clean_pop_rank_df, left_index = True, right_index = True, how= 'inner')
crime_zip_pop_df.reset_index('ZIP', drop=False, inplace=True)

crime_zip_pop_df
```

```python
# Write data to .csv for use in analysis phase
crime_zip_pop_df.to_csv('Resources/clean_pop_zip_model.csv')
```
