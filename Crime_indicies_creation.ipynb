```python
# Dependancies
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import citipy

```

```python
#Loading Files
chicago_pop_zip_data_to_load = os.path.join("Resources","Chicago_Population_by_Zipcode_2020.csv" )
crime_zip_data_to_load = os.path.join("Resources", "Crime_zip.csv")
```

```python
crime_zip_df = pd.read_csv(crime_zip_data_to_load, header=0, delim_whitespace=False)
crime_zip_df
```

```python
clean_crime_df = crime_zip_df[["Date","ZIP", "Primary Type", "rankings"]]
clean_crime_df
```

```python
clean_crime_df =clean_crime_df.rename(columns={"rankings": "crime_severity"})
clean_crime_df
```

```python
#write cleaned data to a .csv
clean_crime_df.to_csv('Resources/cleaned_crime.csv') 
```

```python
# pivot table example with multiple aggregating functions
crime_table3_df = pd.pivot_table(clean_crime_df, values='crime_severity',
                    index='Primary Type',
                     columns='ZIP',
                     aggfunc='sum')
crime_table3_df
```

```python
crime_table2_df = pd.pivot_table(clean_crime_df,  index = 'ZIP', 
                   aggfunc= 'sum')

crime_table2_df
```
