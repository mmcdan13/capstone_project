```python
# Dependencies and Setup
import requests
import gmaps

# Import API key
from config import g_key
    
import pandas as pd
import os
```

```python
crime = pd.read_csv('2020_crime.csv')
crime.head(20)
```

```python
crime = crime.dropna()
crime.head(20)
```

```python
test = crime.sample(n=50000)
test2 = test.copy().reset_index().drop(columns=['index','ID','Case Number','X Coordinate','Y Coordinate','Location','Updated On'])
test2
```

```python
# Set parameters to search for a hotel.
params = {
    "radius": 50000,
    "type": "restaurant",
    "key": g_key
}
```

```python
# Iterate through the DataFrame.
for index, row in test2.iterrows():
    # Get the latitude and longitude.
    lat = row["Latitude"]
    lng = row["Longitude"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    restaurants = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
    test2.loc[index, "Restaurant Name"] = restaurants["results"][0]["name"]
    
    # Grab the first hotel from the results and store the name.
    try:
        test2.loc[index, "Restaurant Name"] = restaurants["results"][0]["name"]
    except (IndexError):
        print("Restaurant not found... skipping.")
```
