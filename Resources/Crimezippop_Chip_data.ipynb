```python
# Dependancies
import pandas as pd
import os
import numpy as np

```

```python
# Loading Files -
crime_model_data_to_load = os.path.join("Resources","clean_pop_zip_model.csv")
chipotle_zip_data_to_load = os.path.join("Resources", 'new_chipotle_data.csv')
```

```python
# reading  data into dataframe for merging
chip_zip_df = pd.read_csv(chipotle_zip_data_to_load,header=0, delim_whitespace=False)
chip_zip_df
```

```python
chip_zip_df['chipotle']=1
```

```python
# reading in data into dataframe for merging

crime_data_df = pd.read_csv(crime_model_data_to_load, header=0, delim_whitespace=False)
crime_data_df
```

```python
crime_data_df.set_index('ZIP', inplace=True)
chip_zip_df.set_index('ZIP', inplace=True)

```

```python

crime_chip_df = pd.merge(crime_data_df, chip_zip_df, on='ZIP', how='outer')
crime_chip_df.reset_index('ZIP', drop=False, inplace=True)

crime_chip_df
```

```python
crime_chip_df_filtered = crime_chip_df[crime_chip_df['rankings'] >=1.0]

crime_chip_df_filtered 
```

```python
crime_chip_df_filtered.dtypes
```

```python


crime_chip_df_filtered['chipotle'] = crime_chip_df_filtered['chipotle'].fillna(0).astype(int)
crime_chip_df_filtered
crime_chip_df_filtered.dtypes
```

```python
crime_chip_df_filtered.drop('Unnamed: 0', axis=1, inplace=True)
crime_chip_df_filtered
```

```python
crime_chip_df_filtered.to_csv('Resources/crime_chip_for_model.csv')
```
